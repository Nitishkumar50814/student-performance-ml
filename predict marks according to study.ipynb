```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
import pandas as pd
data=pd.DataFrame({
    'Hours':[1,2,3,4,5,6,7,8,9],
    'Scores':[15,30,45,50,65,70,85,90,98]
})
```

```python
plt.scatter(data['Hours'], data['Scores'], color='blue')
plt.title('Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

# split the data

```python
x = data[['Hours']] # Feature (  input /Indipendent Variable)
y = data[['Scores']]  # Target( output/dependent variable )
```

```python
x_train, x_test, y_train, y_test = train_test_split (x, y,test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

# predict the Result

```python
y_pred = model.predict(x_test)
print(y_pred)
```

```python
plt.scatter(data['Hours'], data['Scores'], color='red')
plt.title('Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.plot(x_test, y_pred, color='green')
```

```python
print('mean squared error:',mean_squared_error(y_test, y_pred))
print('R2 Score:',r2_score(y_test, y_pred))
```

# predict for costom input

```python
hours = [[7.5]]
predicted_score = model.predict(hours)
print(f'if a student studies for 7.5 hours, predicted score will be:{predicted_score[0][0]:.2f}')
```
